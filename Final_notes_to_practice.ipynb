````python
def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

print(factorial(3))
````

````python
### Visual breakdown of `factorial(3)`

The loop multiplies the running `result` by each integer from 1 to 3:

```
result = 1          # initial value
result = result * 1  -> 1
result = result * 2  -> 2
result = result * 3  -> 6
```

Graphically:

```
1 ──×── 1 ──×── 2 ──×── 3
             ↓          ↓
            1️⃣         2️⃣        = 6️⃣
```

Or as a simple flow:

1 ➜ multiply by 1 ➜ 1
➜ multiply by 2 ➜ 2
➜ multiply by 3 ➜ 6

So `factorial(3)` returns **6**, which is why the print statement outputs **6**.
````

````python
def f(n):
    if n == 0:
        return 1
    return n * f(n-1)
print (f(4))
````

### Visual explanation: Recursive Factorial `f(4)`

This function is **recursive** — it calls itself! The base case stops the recursion when `n == 0`.

**How `f(4)` executes (going down):**

```
f(4)  →  4 * f(3)
           ↓
         3 * f(2)
           ↓
         2 * f(1)
           ↓
         1 * f(0)
           ↓
           1  (base case: return 1)
```

**Then it unwinds (coming back up):**

```
f(0) returns: 1
f(1) returns: 1 × 1 = 1
f(2) returns: 2 × 1 = 2
f(3) returns: 3 × 2 = 6
f(4) returns: 4 × 6 = 24
```

**Result: The print statement outputs `24`**

This is the same factorial you saw before, but written using **recursion** instead of a loop. Each function call remembers what it needs to multiply at the end, then goes deeper until hitting the base case (`n == 0`), then multiplies the results on the way back up.

### Floor Division Operator: `//`

The `//` operator performs **floor division** — it divides and returns only the integer part (drops the decimal).

With `x = 5` and `y = 2`:

| Operation | Result | Meaning |
|-----------|--------|---------|
| `x / y` | 2.5 | Regular division (with decimals) |
| `x // y` | 2 | Floor division (integer only) |

**Visual:**
```
5 ÷ 2 = 2.5
         ↓
       2   (floor division keeps only the integer part)
```

**More examples:**
- `7 // 2` = 3 (7 ÷ 2 = 3.5 → drop the .5)
- `10 // 3` = 3 (10 ÷ 3 = 3.333... → drop the decimal)
- `9 // 3` = 3 (9 ÷ 3 = 3.0 → already an integer)

So your code prints **2** because 5 divided by 2 is 2.5, and floor division rounds down to 2.

### Modulo Operator: `%`

The `%` operator gives you the **remainder** after division. Yes, you're right — it's related to odd/even!

With `5 % 2`:

```
5 ÷ 2 = 2 remainder 1
        ↓
        The % gives you the remainder: 1
```

**Why is it 1? Because 5 is odd!**

| Number | `number % 2` | Meaning |
|--------|-------------|---------|
| 5 | 1 | Odd (remainder 1) |
| 4 | 0 | Even (remainder 0) |
| 7 | 1 | Odd (remainder 1) |
| 6 | 0 | Even (remainder 0) |

**More examples with different divisors:**
- `10 % 3` = 1  (10 ÷ 3 = 3 remainder 1)
- `12 % 5` = 2  (12 ÷ 5 = 2 remainder 2)
- `20 % 4` = 0  (20 ÷ 4 = 5 remainder 0)

**Quick tip:** To check if a number is odd or even:
- If `number % 2 == 1` → **odd**
- If `number % 2 == 0` → **even**


````python
def recursive_count_up(n, odd):
    if n < 0:
        return
    recursive_count_up(n - 1, odd)
    if odd:
        if n % 2 != 0:
            print(n)
    else:
        print(n)

recursive_count_up(5, 1)
````
