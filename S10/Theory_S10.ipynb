else clause exectues code that must be executed if the try clause does nto raise an exception. wHEN THERE'S NO ERRO, EXECUTE THE TRY BLOCK, and the else block will be executed when there's no error, and the finally block will be executed always.

````python
import sys


def divide(x, y):

    try:
        result = x/y

    except ZeroDivisionError:
        sys.stderr.write("Dividing by 0. Returning infinite\n")
        #printint the standard error
        result = float("Inf")

    else:
        print(f"Result is {result}")

    finally:
        print("Executing finally clause")

    return result


print(divide(1, "0"))
````
