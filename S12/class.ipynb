## SUBPROCESS library

```python
import subprocess
subprocess.run(["echo", "Hello, World!"])

# instead of 0 we get CompletedProcess(args=['echo', 'Hello, World!'], returncode=0)
```

## PICKLE

Store a dictionary in a file with pickle

```python
import pickle
# you have a dictionary
my_dict = dict( (a,a+1) for a in range(100) )

# you want to store the dictionary to use it whenever you want
with open("dictionary.pkl", "wb") as f:     # wb: binary
    pickle.dump(my_dict, f)     # we use pickle to store the dictionary in a file called dictionary.pkl
```

Load a dictionary from a file with pickle

```python
with open ("dictionary.pkl", "rb") as f:
    print(pickle.load(f))   # we can load the dictionary from the file and print it
```

- You can only pickle something which all its objects can be pickled too.
- Not very secure as it not control what it saves.
- Is not very portable, only Python.

## JSON library

Store a dictionary in a file with json

```python
import json
my_dict = dict( (a,a+1) for a in range(100) )

json.dumps(my_dict)     # generates a string
with open("dictionary.json", "w") as f:
    json.dump(my_dict, f)
```

Load a dictionary from a file with json

```python
with open("dictionary.json") as f:
    print(json.load(f))
```

- json works with strings so, we have converted the integers to strings
- no tuples in json, it converts to lists
- as the keys are converted to strings, the order is not guaranteed

## URLLIB

```python
import urllib.request
import re

# we can use the urllib library to open a url and read its content
url = urllib.request.urlopen("http://www.elpais.es")

# we can use regular expressions to find all the links in the content of the url
content = url.read()
href_re = re.compile(r"\s+href=\"(http\S+)\"")  # we look for all the href attributes that start with http and we capture the url in a group
all_links = href_re.findall(str(content))
print("\n".join(all_links))
```

# Creating Documentation

In Python the documentation works with Docstrings.

The """ ... """ after the definition of a function is its documentation.

```python
def my_function():
    """ This function does nothing """
    pass

my_function.__doc__  # we can access the docstring of the function with the __doc__ attribute
```

We can also access to the documentation of the function with the help()

```python
help(my_function)  # we can also use the help function to get the docstring of the function
```

You can also have documentation in the module if you put outside the functions a """ ... """

The most simple docstrings are the single-line ones.
 
But there are **multi-line** ones to specify more things.
    In this type of docstrings you can specify the arguments, the return value, etc

## Tools to generate automatically the documentation

```python
import pydoc
pydoc.help(my_function)
```

### Logging

By default, the level of logging is warning so, the debug and info messages will not be seen.

```python
import logging 
logging.debug("This is a debug message")    # we don't see the debug message because the default level of logging is warning
logging.info("This is an info message")   # we don't see the info message because the default level of logging is warning
logging.warning("This is a warning message")    # we see the warning message because the default level of logging is warning
logging.error("This is an error message")   # we see the error message because the default level of logging is warning
```

You can set the level of logging.

```python
import logging
logging.basicConfig(level=logging.DEBUG)   # we change the level of logging to debug
```

```python
logging.debug("This is a debug message")    # we see the debug message because the level of logging is debug
```

# Parallel Processing

- Threading module
  - All the processes share the same memory so, only one execution at a time.
  - You don't gain CPU performance.
- Multiprocessing module
  - Different python executions with separate memory
  - Each job -> a different python process

## Multiprocessing

```python
import multiprocessing
import time

def my_function(tid):
    time.sleep(int(tid) + 10)
    print(f"- Process {tid} finished\n")

procs = 5
# Create a list of jobs and then iterate through
# the number of processes appending each process to
# the job list

jobs = []
for i in range(0, procs):
    process = multiprocessing.Process(target=my_function, args=(i,))
    jobs.append(process)
# Start the processes (i.e. calculate the random number lists)
for j in jobs:
    j.start()
# Ensure all of the processes have finished
for j in jobs:
    j.join()        # wait for all the processes to finish
```
